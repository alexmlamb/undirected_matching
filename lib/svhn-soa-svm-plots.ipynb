```python
import pickle
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
ali = pickle.load(open("stats2-ali.pkl"))
ud = pickle.load(open("stats2.pkl"))
```

```python
zstats_ali = ali[1]
dzstats_ali = ali[0]

zstats_ud = ud[1]
dzstats_ud = ud[0]
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
#plt.plot(np.asarray(dzstats_ud)[:,1], label="Gen Train", linestyle='-', c="blue")
plt.plot(np.asarray(zstats_ali)[:,0], label="ALI Latent Test Error", linestyle='-', c="blue")

#plt.plot(np.asarray(zstats_ud)[:,1], label="Disc Train", linestyle='-', c="red")
plt.plot(np.asarray(dzstats_ali)[:,0], label="ALI Disc-Feat Test Error", linestyle='-', c="red")

#plt.plot(np.asarray(dzstats_ud)[:,1], label="Gen Train", linestyle='-', c="blue")
plt.plot(np.asarray(zstats_ud)[:,0], label="DreamNet Latent Test Error", linestyle='--', c="blue")

#plt.plot(np.asarray(zstats_ud)[:,1], label="Disc Train", linestyle='-', c="red")
plt.plot(np.asarray(dzstats_ud)[:,0], label="DreamNet Disc-Feat Test Error", linestyle='--', c="red")

plt.legend()
```
