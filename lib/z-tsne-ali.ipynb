```python
a = tsne.tsne(np.concatenate((zs[0, ...], zs[500, ...], zs[1000, ...]), axis=0))
a = a.reshape((3, 64, 2))
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (15, 8)
plt.scatter(a[0][:, 0], a[0][:, 1], cmap="plasma", marker="+", label="Step 0")
plt.scatter(a[1][:, 0], a[1][:, 1], cmap="plasma", marker="o", label="Step 500")
plt.scatter(a[2][:, 0], a[2][:, 1], cmap="plasma", marker=".", label="Step 1000")
plt.legend()


# for i, txt in enumerate(coor_tsne):
#     plt.annotate(i, (txt[0],txt[1]))
```
